# Introduction to pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[3,6,9,12],'ro')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--', t,t**2,'bs', t,t**3,'g^') # x, y
# red dashes, blue squares, and green triangles
plt.legend('ABC',ncol=1,loc='upper left')
plt.show()
```

```python
rng = np.random.RandomState(0) # 每次運行後數值皆不變
# rng.random(10)
rng.randn(10)
```

```python
rng = np.random.RandomState(1) # 每次運行後數值皆不變
# rng.randn(10)
rng.random(10)
```

```python
rng = np.random.RandomState() # 每次運行後數值皆改變
# rng.random(10)
rng.randn(10)
```

```python
a = np.random.RandomState(0)
a = a.randint(10, size=10)
a.resize(2,5)
print(a)
np.cumsum(a) # 累加，橫向，攤平
np.cumsum(a,0) # 累加，縱向
np.cumsum(a,1) # 累加，縱向
```

```python
plt.style.use('classic')
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500) # 0~10當中取500個點(Points)，平均分配
# print(x)
rng = rng.randn(500,6) # 500列，6欄 (500, 6)
# print(rng.shape)
y = np.cumsum(rng,0) # 累加，縱向
# print(y)
# print(y.shape)
plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper right')
```

## Scatter Plot: A scatter plot of y v.s. x with varying marker size and/or color

```python
np.arange(50)
np.random.randint(0,50,50)
np.random.randn(50) # from standard normal distribution

# np.arange(50)+ 10*np.random.randn(50)
# 10*np.random.randn(50)
# np.abs(np.random.randn(50))*100
```

```python
data = {'a':np.arange(50),
       'c':np.random.randint(0,50,50),
       'd':np.random.randn(50)}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

print(data['a'])
print(data['c'])

# plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.scatter('a','c',c="c", s='d', data=data) # s: scalar size
plt.xlabel('entry a')
plt.ylabel('entry b')

plt.show()
```
