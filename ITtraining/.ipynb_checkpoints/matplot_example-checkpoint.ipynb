```python
def f(t):
    return np.exp(-t)

t1 = np.arange(0., 5., 0.1)
# print(t1)
t2 = np.arange(0., 5., 0.02)
# print(t2)
plt.figure(1)
plt.subplot(2,1,1)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(2,1,2)
plt.plot( t2, f(t2), 'r--')
plt.show()
```

## Working with Text

```python
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(100000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=2, facecolor='g', alpha=1) # alpha -> 透明度

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(45, 0.025, r'$\mu={0},\ \sigma={1}$'.format(mu, sigma))
plt.axis([40, 160, 0, 0.03])
plt.grid(True) # 是否顯示格線
plt.show()
```

## Annotating Text

```python
ax = plt.subplot(1,1,1)

t = np.arange(0., 5., 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2,1), xytext=(3,1.5), arrowprops=dict(facecolor='black',shrink=0.05))

plt.ylim(-2,2)
plt.show()
```

```python
# data to plot
n_groups = 4
means_frank = (60, 55, 40, 65)
means_guido = (85, 62, 54, 20)
means_total = (145, 117, 94, 85) # (100, 62, 54, 20)

index = np.arange(n_groups) # 4
# print(index) # [0 1 2 3]
bar_width = 0.3
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width, alpha=opacity, color='b', label='Male', align='center')
rects2 = plt.bar(index+bar_width, means_guido, bar_width, alpha=opacity, color='g', label='Female', align='center')
rects3 = plt.bar(index+bar_width*2, means_total, bar_width, alpha=opacity, color='y', label='total', align='center')

plt.xlabel('Education Level')
plt.ylabel('Smoking Percentage')
plt.title('Smoking Percentage vs. Education Level')
plt.xticks(index+bar_width, ('A','B','C','D'))
plt.legend()

plt.tight_layout() # 圖表過度集中可以使用.tight_layout分開

def autolabel(rects):
    '''
    Attach a text label above each bar displaying its height.
    '''
    for rect in rects:
        height = rect.get_height()
        # print(height)
        plt.text(rect.get_x() + rect.get_width()/2, 1.02*height, '%d'%int(height),ha='center',va='bottom')
        
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
plt.ylim(0,160)
plt.show()
```

```python
n_groups = 4
means_frank = (60, 55, 40, 65)
means_guido = (85, 62, 54, 20)
means_total = (145, 117, 94, 85) # (100, 62, 54, 20)

index = np.arange(n_groups) # 4
# print(index) # [0 1 2 3]
bar_width = 0.3
opacity = 0.8

line1 = plt.plot(index, means_frank, alpha=opacity, color='b', marker='o', linestyle='solid', label='Male') # linestyle='solid' -> 實線
for i, j in list(zip(range(0, n_groups), means_frank)): plt.text(i, j+5, j, fontsize=12)    
# print(list(zip(range(0, n_groups), means_frank))) # [(0, 60), (1, 55), (2, 40), (3, 65)]

line2 = plt.plot(index, means_guido, alpha=opacity, color='g', marker='*', linestyle='dashed', label='Female') # linestyle='dashed' -> 虛線
for i, j in list(zip(range(0, n_groups), means_guido)): plt.text(i, j+5, j, fontsize=12) 

line3 = plt.plot(index, means_total, alpha=opacity, color='y', marker='^', linestyle='solid', label='total')
for i, j in list(zip(range(0, n_groups), means_total)): plt.text(i, j+5, j, fontsize=12) 

plt.xlabel('Education Level')
plt.ylabel('Smoking Percentage')
plt.title('Smoking Percentage vs. Education Level')
plt.xticks(index, ('A','B','C','D'))

plt.legend()
plt.tight_layout() # 圖表過度集中可以使用.tight_layout分開

plt.ylim(0,160)
plt.show()
```

```python
# 調節圖形大小、寬、高
plt.figure(figsize=(10,10))
# 定義餅狀圖的標籤，標籤是列表
labels = ['A','B','C','D']

n_groups = 4
means_frank = (60, 55, 40, 65)
means_guido = (85, 62, 54, 20)
means_total = (145, 117, 94, 85) # (100, 62, 54, 20)

vec_total = [x+y+z for x,y,z in zip(means_frank, means_guido, means_total)]
# print(list(zip(means_frank, means_guido, means_total))) # [(60, 85, 145), (55, 62, 117), (40, 54, 94), (65, 20, 85)]

# 每個標籤占多大，會自動去算百分比
sizes = vec_total
colors = ['orange', 'red', 'yellowgreen', 'lightskyblue']

# 將某部分爆炸出來，使用括弧，將第一塊分割出來，數值的大小是分割出來的與其他兩塊的間隙
explode = (0.05, 0, 0, 0)

patches, l_text, p_text = plt.pie(sizes, explode=explode, labels=labels, colors=colors, labeldistance=1.1, autopct='%3.1f%%', 
                                 shadow=True, startangle=90, pctdistance=0.6) 

# labeldistance: 文本的位置離原點有多遠，1.1指1.1倍半徑的位置
# autopct: 圖裡面的文本格式，'%3.1f%%'表示小數有三位元，整數有一位元的浮點數
# shadow: 餅是否有陰影
# startangle: 起始角度，0表示從0開始逆時針轉，為第一塊，一般選擇從90度開始比較好看
# pctdistance: 百分比的text離圓心的距離
# patches, l_text, p_text，為了得到圓形圖的返回值，p_text圓形圖內部文本的，l_text圓形圖外label的文本
# print(patches) # print(p_text) # print(l_text)

# 改變文本的大小
for t in l_text: t.set_size(15)
for t in p_text: t.set_size(12)
    
# 設置x,y軸刻度一致，這樣圓型形才能是圓的
plt.axis('equal')
plt.title('Smoking Percentage vs. Education Level')
# plt.legend()
plt.show()
```

# Seaborn

```python
import seaborn as sns
sns.set()
x = np.array([1,2,3,4])
y = np.power(x, 3) # 三次方
plt.plot(x, y)
plt.legend('ABCDEF', ncol=1, loc='upper left')
```

```python
import pandas as pd
data = np.random.multivariate_normal([0,0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x','y'])
print(data.shape)

for col in 'xy':
    plt.hist(data[col], alpha=0.5) #normed=True
```

```python
# %matplotlib inline 可以在Ipython编译器里直接使用，功能是可以内嵌绘图，并且可以省略掉plt.show()这一步。
%matplotlib inline 
df_iris = pd.read_csv('iris.csv')
# print(df_iris)
fig, axes = plt.subplots(1, 2)
sns.distplot(df_iris['petal length (cm)'], ax=axes[0], kde=True, rug=True) # kde: 密度曲線，核密度估計（Kernel density estimation）; rug: 在每個觀察點上的垂直小標籤
sns.kdeplot(df_iris['petal length (cm)'], ax=axes[1],  shade=True) #shade: 陰影
```

## pairplot: plot pairwise relationships in dataset.
