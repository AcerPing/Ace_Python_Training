```python
# 匯入所需函示庫
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 準備輸出資料
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
```

```python
# 描繪圖表
fig, ax = plt.subplots()
ax.plot(t, s)
```

```python
# 繪製圖表的標籤與格線
ax.set(xlabel='time(s)', ylabel= 'voltage(mV)', title = 'About as simple as it gets, folks')
ax.grid()
```

```python
# 顯示繪製圖表
plt.show()
```
